# 1. Membuat Simulasi Data

## 1.1 Simulasi Data Ibu Hamil

Dalam panduan analisis data antropometri berikut ini akan digunakan data simulasi yang terdiri atas 100 observasi. Data simulasi diambil berdasarkan data rata-rata ibu hamil yang diperbanyak secara acak menggunakan distribusi normal menggunakan library Numpy. Data simulasi ini terdiri atas 5 variabel, yaitu tinggi badan, panjang lingkar lengan, sistolik, diastolik, dan IMT.

```python
#Konstanta di bawah ini hanya digunakan untuk mendesain sebaran data. Dapat diubah sesuai keinginan

std_a=3 #Mengatur simpangan baku tinggi badan
std_b=4 #Mengatur simpangan baku lingkar badan
std_sis=20 #Mengatur simpangan baku tekanan darah (sistolik)
std_dias=30 #Mengatur simpangan baku tekanan darah (diastolik)
std_imt=7.5 #Mengatur simpangan baku IMT

mean_a=150 #Mengatur rerata tinggi badan
mean_b=21.5 #Mengatur rerata lingkar lengan
mean_sis=120 #Mengatur rerata tekanan darah (sistolik)
mean_dias=80 #Mengatur rerata tekanan darah (sistolik)
mean_imt=21.7 #Mengatur rerata  IMT

tinggi_bdn_ih=mean_a+ std_a*np.random.randn(100) #simulasi data tinggi badan ibu hamil dalam cm 
lingkar_lgn_ih=mean_b+ std_b*np.random.randn(100) #simulasi data lingkar lengan ibu hamil dalam cm
sis_ih=mean_sis+ std_sis*np.random.randn(100)
dias_ih=mean_dias+ std_dias*np.random.randn(100)
imt_ih=mean_imt+std_imt*np.random.randn(100)

df_ih=pd.DataFrame({'tinggi_badan':tinggi_bdn_ih,
                   'lingkar_lengan': lingkar_lgn_ih,
                   'sistolik':sis_ih,
                    'diastolik':dias_ih,
                    'IMT':imt_ih})
df_ih
```

# 2.1 Statistika Deskriptif

## 2.1.1 Sebaran data dan pusat data

```python
df_ih.describe() #Seberan data: std, kuartil, min, maks ; Pusat data: mean, median
```

# 3. Uji kenormalan data

## 3.1 Q-Q plot

```python
import pylab
import scipy.stats as stats
```

```python
for variabel in df_ih.columns:
    stats.probplot(df_ih[variabel], dist="norm", plot=pylab)
    plt.title('Plot QQ untuk '+variabel)
    pylab.show()
```

# 3.2 Shapiro Wilk Test

```python
from scipy.stats import shapiro
```

```python
for variabel in df_ih.columns:
    print(variabel)
    print(shapiro(df_ih[variabel]))
    print('-----')
```

# 3.3 Kolmogorov-Smirnov Test

```python
from scipy.stats import kstest
```

```python
for variabel in df_ih.columns:
    print(variabel)
    print(kstest(df_ih[variabel], 'norm'))
    print('-----')
```
